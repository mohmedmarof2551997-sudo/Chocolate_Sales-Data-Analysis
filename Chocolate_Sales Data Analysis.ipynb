<h2 style="color:#2E86C1;">
5️⃣🧹Data Cleaning & Preprocessing
</h2>

💡 :
The Amount(sales) column was originally stored as text with non-numeric characters and was cleaned and converted into a numeric data type.

Converted the Date column into datetime format.

Extracted time-based features: Month \Year\ Quarter

Created a new column Revenue_per_Box to measure product return efficiency.

```python
# Clean Amount column
df['Amount'] = df['Amount'].str.replace('[$,]', '', regex=True).str.strip().astype(float)
```

```python
print(f"Column Amount after cleaning:\n{df['Amount'].head()}")
```

```python
#dates (mixed formats) COVERT TO DATETIME

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df['Year']           = df['Date'].dt.year
df['Month']          = df['Date'].dt.month
df['MonthName']      = df['Date'].dt.strftime('%b')
df['YearMonth']      = df['Date'].dt.to_period('M')
df['Quarter']        = df['Date'].dt.quarter
```

```python
df
```

```python
#create new column that help at analysis (Revenue_per_Box)

df['Revenue_per_Box'] = df['Amount'] / df['Boxes Shipped']
```

1) What is the total sales value and total boxes shipped for the selected period?

2) What is the month-over-month growth rate in total sales?

```python
monthly = df.groupby('YearMonth')['Amount'].sum()
mom = monthly.pct_change().dropna()* 100

print(f"Average Month-over-Month Growth: {mom.mean():.2f}%")
print(f"Best Month  : {monthly.idxmax()} (${monthly.max()})")
print(f"Worst Month : {monthly.idxmin()} (${monthly.min()})")
```

```python
plt.figure(figsize=(14,7))
plt.plot(mom.index.astype(str), mom.values, marker='o', color= 'orange' , label='MoM Growth (%)')
plt.scatter(str(mom.idxmax()), mom.max(), color='g', s=100, label='Max Growth')
plt.scatter(str(mom.idxmin()), mom.min(), color='red', s=100, label='Min Growth')
plt.xlabel("Year-Month",fontsize=16,labelpad=10)
plt.ylabel("Month-over-Month Growth (%)",fontsize=16,labelpad=10)
plt.title("Month-over-Month Growth (%)",fontsize=16,pad=10)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

💡 Insight : 

1)+0.75% Monthly groth average,

January is always the strongest, Best month = January 2024 ← $1,013,621

2)April is historically the weakest month ,Worst month = April 2022 ← $674,051

3)The company is improving

April 2022 → $674,051 ← Worst month

January 2024 → $1,013,621 ← Best Month

Growth = Total +50% over the past years

3) What is the year-over-year growth rate in sales by month?

```python
yearly = df.groupby('Year')['Amount'].sum()
growth = yearly.pct_change().dropna() * 100
print(f"Average year over year Growth: {growth.mean():.2f}%")
print(f"Best Year  : {growth.idxmax()} (${yearly.max()})")
print(f"Worst year : {growth.idxmin()} (${yearly.min()})")
print(f"Best Year  (Growth)  : {growth.idxmax()} ({growth.max():+.1f}%)")
print(f"Worst Year (Growth)  : {growth.idxmin()} ({growth.min():+.1f}%)")
growth.plot(kind='line', marker='o', figsize=(10, 5), 
            title='GROWTH YEAR OVER YEAR', color='brown', linewidth=3)
plt.ylabel('YOY GROWTH (%) ',labelpad=10)
plt.xlabel('Year',labelpad=10,fontsize=15)
plt.axhline(growth.mean(), color='orange', linestyle='--', label=f'AVERAGE = {growth.mean():.1f}%')
plt.axhline(0, color='gray', linestyle='-', alpha=0.5)
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

4)Which country contributes the highest share of total revenue?

```python
country_rev = df.groupby('Country').agg(
    Revenue=('Amount','sum'),
    Boxes=('Boxes Shipped','sum'),
    Transactions=('Amount','count')
).sort_values('Revenue', ascending=False)
country_rev['Revenue Share %'] = (country_rev['Revenue']/country_rev['Revenue'].sum()*100).round(2)
country_rev
```

```python
ax = country_rev['Revenue Share %'].plot(kind='bar', y='Country', figsize=(10,5))
for bar in ax.containers:
    ax.bar_label(bar, fmt='%.1f%%')  
plt.title("COUNTRY REVENUE SHARE")
plt.ylabel("Revenue Share %")
plt.xlabel("Country")
plt.show()
```

💡 Insight : 
    
✅Australia is the most profitable market → expansion opportunity.

✅ India & USA have high transaction volume → improved pricing could boost revenue.

✅ Canada ships a lot but generates less revenue → high potential for discounts.

✅ New Zealand has consistent performance → stable, low-risk market.

7)What is the average revenue per box overall and by country?

```python
rpb = df.groupby('Country')['Revenue_per_Box'].agg(['mean','min','max'])
rpb.columns = ['Avg Rev/Box','Min','Max']
rpb = rpb.sort_values('Avg Rev/Box', ascending=False)
print("Revenue per Box by Country:")
rpb
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(
    data=rpb,
    x='Country',
    y='Avg Rev/Box',
    palette='viridis'
)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')

plt.title("Average Revenue per Box by Country", fontsize=16, weight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Avg Revenue per Box ($)", fontsize=14)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

💡 Insight :

Canada and USA → Premium markets, high revenue per fund, worth focusing on profitable products.

UK and India → Low revenue per fund, but there are some High Value Boxes → Upsell opportunities.

Australia and New Zealand → Average revenue is good

8) How does product sales mix change over time?

```python
top_5 = df.groupby('Product')['Amount'].sum().nlargest(5).index.tolist()
product_comparison = df[df['Product'].isin(top_5)].groupby(['Year', 'Product'])['Amount'].sum().unstack()
ax=product_comparison.plot(kind='bar', figsize=(14, 6), width=0.8,colormap='viridis')
plt.title('Comparing sales of top products over the years', fontsize=14, fontweight='bold')
plt.xlabel('year')
plt.ylabel('revenue ($)')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', fontsize=10)
plt.legend(title='Product',
           bbox_to_anchor=(1.02,1))
plt.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

9) What seasonal patterns exist across months and quarters?

```python
monthly_sales = df.groupby(['Month','MonthName'])['Amount'].sum().reset_index()
print(monthly_sales)
plt.figure(figsize=(12,6))

sns.lineplot(
    data=monthly_sales,
    x='MonthName',
    y='Amount',
    marker='o'
)

plt.title("Monthly Sales Seasonality", fontsize=15, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)

plt.grid(True, alpha=0.3)
plt.show()
```

```python
quarter_sales = df.groupby(['Year','Quarter'])['Amount'].sum().reset_index()
print(quarter_sales)
plt.figure(figsize=(12,6))

ax=sns.barplot(
    data=quarter_sales,
    x='Year',
    y='Amount',
    hue='Quarter',
    palette='viridis'
)
for container in ax.containers:
    ax.bar_label(container,
                 fmt='$%.0f',
                 padding=3,
                 fontsize=10,
                 weight='bold')

plt.title("Quarterly Sales Performance", fontweight='bold')
plt.xlabel("year")
plt.ylabel("Revenue ($)")
plt.legend(title='Quarter')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
pivot_q = quarter_sales.pivot(index='Year', columns='Quarter', values='Amount')

plt.figure(figsize=(8,5))

sns.heatmap(
    pivot_q,
    annot=True,
    fmt='.0f',
    cmap='YlGnBu'
)

plt.title("Quarterly Revenue Seasonality Heatmap", fontweight='bold')
plt.ylabel("Year")
plt.xlabel("Quarter")

plt.show()
```

💡 Insight :

Every year without exception
Q1 → Strongest ✅
Q3 → Weakest ⚠️

Quarter 3 is the weakest quarter 📉

A clear decline year-on-year.

Almost 30% lower than Q1.

Possible reasons:

Holiday season

End of sales cycle

⚠️ Very clear seasonal drop
    




10 )What percentage of total revenue comes from the top 5 products?

```python
top5 = (
    df.groupby('Product')['Amount']
      .sum()
      .nlargest(5)
)

percentage = top5.sum() / df['Amount'].sum() * 100

print(top5)
print(f"\nTop 5 Products Contribution: {percentage:.2f}%")
```

```python
top5 = (
    df.groupby('Product')['Amount']
      .sum()
      .nlargest(5)
)
others = df['Amount'].sum() - top5.sum()
pie_data = top5.copy()
pie_data['Others'] = others

plt.figure(figsize=(8,8))

plt.pie(
    pie_data,
    labels=pie_data.index,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85
)


plt.tight_layout()
plt.show()
```

💡 Insight :

The top 5 products account for only 26.8% of profits.

This means the remaining 17 products make up 73.2%.

The portfolio is very well-balanced. 
    
It doesn't rely on just one product.
